#### Создание признакового простанства с помощью библиотеки sklearn 

```python
# bag_of_words for stemming data of tweeter
count_vectorizer_stemmed = CountVectorizer(ngram_range=(1, 1),
                                   analyzer='word',
                                   max_df=0.9,
                                   stop_words="english",
                                   max_features=1000,
                                   binary=False,)

count_vectorizer_stemmed.fit(list(chain.from_iterable(tweets['tweet_stemmed'].tolist())))
bag_of_words_stemmed = count_vectorizer_stemmed.fit_transform(tweets['tweet_stemmed'].apply(lambda x: ' '.join([word for word in x])))
feature_names = count_vectorizer_stemmed.get_feature_names()
df_stemmed = pd.DataFrame(bag_of_words_stemmed.toarray(), columns = feature_names)
df_stemmed
```

```python
#find any word in first tweet
in_dict = [tweet for tweet in tweets['tweet_stemmed'][0] if tweet in count_vectorizer_stemmed.get_feature_names()]
in_dict
```

```python
# check if word 'father' get in matrix in first row
df_stemmed['father'][0]
```

### Обучение модели Word2Vec

```python
from gensim.models import Word2Vec
model_tweets = Word2Vec(size=200,
                       window = 5,
                       min_count = 2,
                       sg = 1,
                       hs = 0,
                       negative = 10,
                       workers = 32,
                       seed = 34)
```

```python
# create vocabular
model_tweets.build_vocab(tweets['tweet_token'])
```

```python
model_tweets.train(tweets['tweet_token'], total_examples=len(tweets['tweet_token']), epochs=20)
```

```python
# find similary word for word 'dinner'
result = model_tweets.wv.most_similar(positive=['user'])
print(result)
```

```python
# find similary word for word 'trump'
result = model_tweets.wv.most_similar(positive=['trump'])
print(result)
```

```python
# get vector for word 'food'
result = model_tweets.wv.get_vector('food')
print(result)
```

```python
vector_w2v = np.zeros((len(tweets['tweet_token']), 200))
for i, tweet in enumerate(tweets['tweet_token']):
    for t in tweet:
        vector_w2v[i] += model_tweets.wv[t]
    l = len(tweet)
    if l > 0:
        vector_w2v[i] /= l

vector_w2v
```
